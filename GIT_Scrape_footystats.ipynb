### Scrape of Footystats.org CSV file ID's using Selenium to manipulate login credentials
https://footystats.org/download-stats-csv
#### The footystats.org website is a comprehensive subscription based soccer data repository. This code demonstrates a Selenium library based python script to overcome login/password requirements to access its API and retrieve chosen csv files, organize the data to user requirements and save to a specific named excel sheet for further processing. This is an example of a first stage data collection and transformation procedure for a soccer auto pricing strategy.

```python
# Import required libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select  
import pandas as pd
import time

pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None  

USERNAME = '..add_username_here..'
PASSWORD = '..add_passwrod_here..' 
# Insert the path of chromedriver, Selenium works best via Chrome browser (downloaded from https://sites.google.com/a/chromium.org/chromedriver/downloads)
CHROMEDRIVER_PATH = 'C:\\Users\\Documents\\Python\\chromedriver.exe' 
driver = webdriver.Chrome(executable_path=CHROMEDRIVER_PATH)  
driver.get('https://footystats.org/login');
search_box = driver.find_element('name','username') 
search_box.send_keys(USERNAME)
search_box = driver.find_element('name','password') 
search_box.send_keys(PASSWORD)
driver.find_element('id','register_account').submit() 

# Downlaod Premier league season matches for our sample
matches = driver.get('https://footystats.org/c-dl.php?type=matches&comp=7704'); 

#Access the now downloaded file and convert to a data frame
df = pd.read_csv('C:\\Users\\Downloads\\england-premier-league-matches-2022-to-2023-stats.csv')

driver.quit()
```

```python
# Store the last round of fixtures with market prices for analysis
# firstly isolate the fixtures in a new df
last_round = df_complete['GW'].max()
df_last_round = df_complete[df_complete['GW'] == last_round]
df_last_round
```

```python
# save last round to same excel workbook, different sheet
with pd.ExcelWriter('C:\\Users\\Documents\\Auto_Pricer_Prem_22_23.xlsx', engine="openpyxl", mode="a",if_sheet_exists="overlay") as writer:
    df_last_round.to_excel(writer, sheet_name="Import_Last_Rnd", index=False)
print('Saved successfully')
```

```python
# Store next round of matches for next stage pricing. Isolate matches with rows status'incomplete' 
# create new df just with matches yet to be played
df_incomplete = df[df['status']=='incomplete']
```

```python
# create variable 'next_round' consisting of the Game Week number located at index [0,4] for the next set of matches
# We want the next two rounds 
next_round = df_incomplete.iloc[0,4]
df_next_rounds1 = df_incomplete[(df_incomplete['GW'] == next_round)] 
df_next_rounds2 = df_incomplete[(df_incomplete['GW'] == next_round +1 )]
```

```python
# combine the rounds
df_next_rounds = pd.concat([df_next_rounds1,df_next_rounds2])
# isolate only the columns we want
df_next_rounds = df_next_rounds[['date_GMT','Home','Away']]
df_next_rounds
```

```python
# Save to the the same excel workbook, different sheet
with pd.ExcelWriter('C:\\Users\\Documents\\Auto_Pricer_Prem_22_23.xlsx', engine="openpyxl", mode="a",if_sheet_exists="overlay") as writer:
    df_next_rounds.to_excel(writer, sheet_name="Import_Next_Rnd", index=False)
print('Saved successfully')
```
